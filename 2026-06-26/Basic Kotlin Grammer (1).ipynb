## 함수의 기본 값 정의
-함수 인수에 값을 지정하면, 가변 인수를 전달할 수 있음

```kotlin
import kotlin.math.pow   
fun power(base: Double, exp: Int = 2): Double {
    return base.pow(exp)
}

println("2^3 = ${power(2.0, 3)}")
println("3^2 = ${power(3.0, 2)}")
```

## when 제어문 사용
-when 제어문은 기존의 switch 문보다 강력함
-일반적인 switch문은 if 동등절에 해당하나, when 제어문은 범위 및 다양한 조건을 포함할 수 있음

```kotlin
//println("입력 스트림")
val input: String = readln()
val data : Any = input
when(data) {
    is String -> println("입력이 String임")
    20, 30 -> println("입력 값이 20 또는 30임")
    in 1..10 -> println("입력 값이 1이상 10이하")
    else -> println("입력 값이 예상 범위를 벗어남")
}
```

```kotlin
fun series1(n: Int): Double {
    var sum = 0.0

    for (i in 0 until n) {
        val denominator = 2 * i + 1
        val term = 1.0 / denominator

        if (i % 2 == 0) {
            sum += term
        } else {
            sum -= term
        }
    }

    return sum * 4
}

println(series1(100))
println(series1(1000))
println(series1(5000))
```

## 반복문을 통해서 원주율 계산하기
-pi/4 = 1 - 1/3 + 1/5 - 1/7 ...

```kotlin
fun series(n: Int): Double {
    var x: Double = 0.0
    for (i in 1..n) {
        val denom = 2 * i - 1
        val sign = if (i % 2 == 1) 1 else -1
        x += sign / denom.toDouble()
    }
    return x * 4
}

println(series(100))
println(series(1000))
println(series(5000))
```
